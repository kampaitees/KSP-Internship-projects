# Facial Landmarks Detection using Dlib

```python
#importing the necessary dependencies
import cv2
import dlib
import numpy as np
from imutils import face_utils
import matplotlib.pyplot as plt

#creating the detector object for detecting the face from the input image
detector = dlib.get_frontal_face_detector()

#creating the predictor object for  preciting landmarks of face using dlib's pre-trained models 
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

```python
#reading the input image using opencv
image = cv2.imread('images/jerry_seinfeld/7_a.jpg')

#as images are in BGR format in OpenCV while in Matplotlib images are in RGB format so converting them in RGB format
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#seeing the input image
plt.imshow(image_rgb)
plt.show()
```

```python
#creating the copy of input image so that input image can be used later
image_rgb_copy = image_rgb.copy()

#detecting the face from the input image of person
landmarks = detector(image_rgb, 0)

#predicting the landmarks on the face detected image
for landmark in landmarks:
    
    shape = predictor(image_rgb, landmark)
    
    #converting the image into numpy array
    shape = face_utils.shape_to_np(shape)
    
    #drawing the landmarks on the copy of the original image
    for x, y in shape:
        
        cv2.circle(image_rgb_copy, (x, y), 2, (0, 255, 255), -1)

#let see our output result
plt.imshow(image_rgb_copy)
plt.show()
```
